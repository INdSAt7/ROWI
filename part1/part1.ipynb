```python
def generate_date_range(start_date: date, end_date: date) -> list[date]:
    return tuple([(start_date + timedelta(days=i)).strftime('%Y-%m-%d') for i in range((end_date - start_date).days + 1)])

def date_processing(dates: str) -> tuple[date, date]:
    start_str, end_str = dates.split('-')
    start_date = datetime.strptime(start_str, "%d.%m.%Y").date()  
    end_date = datetime.strptime(end_str, "%d.%m.%Y").date()
    
    return generate_date_range(start_date, end_date)
```
